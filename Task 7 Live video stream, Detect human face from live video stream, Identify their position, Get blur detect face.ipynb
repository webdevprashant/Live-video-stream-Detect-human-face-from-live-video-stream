```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
```

```python
ret , photo = cap.read()
```

```python
model = cv2.CascadeClassifier('nov-12-haarcascade_frontalface_default-Copy1.xml')
```

```python
fdetect = model.detectMultiScale(photo)
fdetect
```

```python
while True:
    ret , photo = cap.read()
    fdetect = model.detectMultiScale(photo)
    if len(fdetect) == 0:
        print("No Faces Found")
    else:
        x1 = fdetect[0][0]
        y1 = fdetect[0][1]
        x2 = fdetect[0][2] + x1
        y2 = fdetect[0][3] + y1
        rphoto = cv2.rectangle(photo , (x1,y1) , (x2,y2) , [0,255,0] , 2)
        cphoto = rphoto[y1:y2 , x1:x2]
        blurimg = cv2.blur(cphoto , (20 , 20))
        photo[y1:y2,x1:x2] = blurimg
        cv2.imshow("hi" , photo)
        if cv2.waitKey(10) == 13:
            break
cv2.destroyAllWindows()
```

```python
cap.release()
```
